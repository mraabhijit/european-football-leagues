```python
# "https://api.football-data.org/v4/competitions/PL/teams"

base_uri = "https://api.football-data.org/v4/competitions/"

teams = {"PL": "Premier League", 
        "BL1": "Bundesliga",
        "FL1": "Ligue 1",
        "PD": "La Liga",
        "SA": "Serie A"}

team_dict = {}

for key, value in teams.items():
    uri = base_uri + key + '/teams'

    print(f"Getting data for {value} teams")
    try: 
        response = requests.get(uri, headers=headers)
        print(response)
        print()
        team_dict[value] = response.json()['teams']
    except FileNotFoundError:
        print("Data not available at {uri}")
```

```python
league_teams = {"Premier League": 20, 
        "Bundesliga": 18,
        "Ligue 1": 18,
        "La Liga": 20,
        "Serie A": 20}
```

```python
df = pd.DataFrame(columns = ['team_id', 'team', 'full_name', 'abbv', 
                             'team_crest', 'address', 'country', 'year_founded', 
                             'stadium', 'coach', 'coach_nationality'])

for league, value in league_teams.items():
    print(f"Adding teams of {league}...")
    # print("-"*50)
    for i in range(value):
        data = team_dict[league][i]
        team_id  = data['id']
        team = data['shortName']
        full_name = data['name']
        abbv = data['tla']
        team_crest = data['crest']
        address = data['address']
        country = data['area']['name']
        year_founded = data['founded']
        stadium = data['venue']

        # if data['coach']['lastName'] != '':
        #     coach = (data['coach']['firstName'] + ' ' + data['coach']['lastName']).strip()

        if data['coach']['lastName'] == '':
            coach = data['coach']['firstName']
        elif data['coach']['firstName'] == '':
            coach = data['coach']['lastName']
        print(team, coach)
        coach_nationality = data['coach']['nationality']

        df.loc[len(df.index)] = [team_id, team, full_name, abbv, team_crest, address, country, year_founded, stadium, coach, coach_nationality]
        # print(team_id, team, full_name, abbv, team_crest, address, country, year_founded, stadium, coach, coach_nationality)
    # print("-"*50)
    # print()
```
